```python
import pandas as pd
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("ExcelReader").getOrCreate()

url = "https://www.dol.gov/sites/dolgov/files/ETA/oflc/pdfs/LCA_Disclosure_Data_FY2024_Q4.xlsx"
df_pandas = pd.read_excel(url)

# 2) Convert pandas DataFrame -> Spark DataFrame
df_spark = spark.createDataFrame(df_pandas)

df_spark.show()
df_spark.printSchema()
```
