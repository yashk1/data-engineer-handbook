```python
matches_bucketed = spark.read \
                    .format('parquet') \
                    .load('/home/iceberg/notebooks/notebooks/src/jobs/spark-warehouse/matches_bucketed')

matchdetails_bucketed = spark.read \
                    .format('parquet') \
                    .load('/home/iceberg/notebooks/notebooks/src/jobs/spark-warehouse/matchdetails_bucketed')

medals_matches_players_bucketed = spark.read \
                    .format('parquet') \
                    .load('/home/iceberg/notebooks/notebooks/src/jobs/spark-warehouse/medals_matches_players_bucketed')
```

```python


final_df = matches_bucketed \
            .join(matchdetails_bucketed, on='match_id') \
            .join(medals_matches_players_bucketed, on=['match_id','player_gamertag'])
final_df.show()
```

```python
from pyspark.sql import functions as F

final_df.groupby('match_id', 'player_gamertag') \
        .agg(F.avg('player_total_kills').alias('avg_kills_per_game')) \
        .orderBy(F.desc('avg_kills_per_game')).show(1)
```

```python
final_df.groupby('playlist_id') \
        .agg(F.countDistinct('match_id').alias('matches')) \
        .orderBy(F.desc('matches')).show(1)
```
